```python

#ast is a library and literall_eval is a function in ast, it converts the string into a list of string that can be used in the for loop to provide the required output
import ast
def change(obj):    
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

```

```python
#calling the helper function and applying the change in the genres sectiom
movies['genres']= movies['genres'].apply(change)
```

```python
movies.head()
```

```python
#doing the same thing with keywords

movies['keywords'] = movies['keywords'].apply(change)
```

```python
movies.head()
```

```python
#now we need the top 4 names from the cast so we put a counter =0 and in the loop initialized an if else condition to break the loop after scanning the 4 cast members
import ast
def convert3(obj):    
    counter = 0
    L = []
    for i in ast.literal_eval(obj):
        if counter <=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L

```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
#in crew section we only need the name of the director so well check the job of each crew and break the loop once we found the name of the director
import ast
def fetch_director(obj):    
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
     
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
#to slpit every word and make it a single identity
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# here we will remove all spaces from the names by using a function called replace(" ","")
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" " , "") for  i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" " , "") for  i in x])
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" " , "") for  i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" " , "") for  i in x])
```

```python
movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
#stemming is a process which converts words into there root form
from nltk.stem.porter import PorterStemmer

def stem(text):
    ps = PorterStemmer()
    # Check if input is a list
    if isinstance(text, list):
        text = " ".join(text)  # Join the list into a single string
    words = text.split()  # Now it's safe to call split()
    stemmed_words = [ps.stem(word) for word in words]
    return " ".join(stemmed_words)
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df
```

```python
new_df['tags'] = new_df['tags'].apply (lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#converted all the characters into lower space as it is recommended
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
#sklearn(SicKit learn) is an python library
#CountVectorizer is a class which is used to convert text data into numerical vectors

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000 , stop_words= 'english')
```

```python
# text data that needs to be converted into numerical representations suitable for machine learning algorithms. 
#This process is known as feature extraction.
#this is done with the help of fit_transform which is used for vectorised any text data

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=1000 , stop_words= 'english')

vectors = cv.fit_transform(new_df['tags'])
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title']==  movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted (list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:11]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
    

```

```python
from sklearn.neighbors import NearestNeighbors

#  Initialize k-NN model
knn = NearestNeighbors(metric='cosine', algorithm='brute')

#  Fit the model with the vector data
knn.fit(vectors)

# Function to recommend movies using k-NN
def recommend_knn(movie_title, k=5):
    movie_index = new_df[new_df['title'] == movie_title].index[0]
    
    # Find k nearest neighbors
    distances, indices = knn.kneighbors(vectors[movie_index], n_neighbors=k+1)  # Include the movie itself
    
    #  Retrieve and display the recommended movies
    recommended_movies = []
    for idx in indices.flatten()[1:]:  # Skip the first, as it's the input movie
        recommended_movies.append(new_df.iloc[idx].title)
    
    return recommended_movies


print(recommend_knn("Avatar"))
```
